```python
import numpy as np
import pandas as pd
```

```python
titanic_train = pd.read_csv('https://raw.githubusercontent.com/training-ml/Files/main/titanic_train.csv')
```

```python
titanic_train.head()
```

```python
#how to check unique values of any column
titanic_train['Survived'].unique()   #Similar you can check any column unique values
```

```python
#how to rename categories using pandas method
new_survived = pd.Categorical(titanic_train['Survived'])
new_survived = new_survived.rename_categories(['Died','Survived'])
new_survived.describe()
```

```python
titanic_train['Pclass'].unique()
```

```python
new_Pclass = pd.Categorical(titanic_train["Pclass"],
                            ordered=True)
new_Pclass = new_Pclass.rename_categories(["Class1","Class2","Class3"])
new_Pclass.describe()
```

```python
# Reassign New name to Pclass
titanic_train['Pclass'] = new_Pclass
```

```python
titanic_train.head()
```

```python
titanic_train['Survived']=new_survived
```

```python
titanic_train.head()
```

```python
titanic_train['Cabin'].unique()
```

```python
char_chain = titanic_train['Cabin'].astype(str)   #Convert Data to Str

# list Comprehension (Creating list using another set of string, list etc)

new_Cabin= np.array([cabin[0] for cabin in char_chain])  #Take first letter
new_Cabin = pd.Categorical(new_Cabin)
new_Cabin.describe()
```

```python
titanic_train['Cabin']= new_Cabin
```

```python
titanic_train.head(10)
```

```python
#find all passengers who's age is missing
## np.where gives the index of thosse missing data
missing= np.where(titanic_train['Age'].isnull()==True)   # find all nulls is column age
missing
#missing.length()
```

```python
#using the index, locate passangers
titanic_train.loc[missing]
```

```python
#find the maxmium fare
index = np.where(titanic_train['Fare']==max(titanic_train['Fare']))
index
```

```python
titanic_train.loc[index]
```

```python
#find family who has hightest members
titanic_train['Family']= titanic_train['SibSp']+titanic_train['Parch']
most_family = np.where(titanic_train['Family']==max(titanic_train['Family']))
most_family
```

```python
titanic_train.loc[most_family]
```
